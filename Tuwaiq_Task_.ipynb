# Training:
This section focuses on the modeling process and compares the performance of Random Forest, Logistic Regression, and Ensemble Learning techniques. The main objective was to identify the most effective approach for our problem.
After evaluating the performance of Random Forest, Logistic Regression, and Ensemble Learning models, the Ensemble Learning approach achieved the highest result. The GridSearchCV further enhanced the performance by finding the optimal hyperparameters for the ensemble model.

Random Forest:

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_predict,train_test_split
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import classification_report

# Split the dataset into training and test sets
X = processed_df.drop('Y', axis=1)
y = processed_df['Y']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)

# Apply SMOTE for dealing with class imbalance
sm = SMOTE(random_state=42)
X_train_res, y_train_res = sm.fit_resample(X_train, y_train)

# Logistic Regression Classifier
classifier = RandomForestClassifier()

# Define hyperparameters for RandomizedSearchCV
param_grid = {'max_depth': [10],
              'min_samples_leaf': [4],
              'min_samples_split': [2, 5, 10],
              'n_estimators': [400, 1000]}

# Perform RandomizedSearchCV
rf_random2 = RandomizedSearchCV(estimator=classifier, param_distributions=param_grid, scoring='f1', n_iter=100, cv=5, verbose=2, random_state=42, n_jobs=-1)
rf_random2.fit(X_train_res, y_train_res)

# Get the best model from RandomizedSearchCV
best_rf_model = rf_random2.best_estimator_

# Make predictions on the training and test sets
y_train_pred = best_rf_model.predict(X_train_res)
y_test_pred = best_rf_model.predict(X_test)

# Evaluate the model on both training and test sets
train_accuracy = accuracy_score(y_train_res, y_train_pred)
test_accuracy = accuracy_score(y_test, y_test_pred)

print("Training Accuracy:", train_accuracy)
print("Test Accuracy:", test_accuracy)

# Generate classification reports for both training and test sets
print("Training Classification Report:")
print(classification_report(y_train_res, y_train_pred))

print("Test Classification Report:")
print(classification_report(y_test, y_test_pred))
```

Logistic Regression:

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Split the dataset into training and test sets
X = processed_df.drop('Y', axis=1)
y = processed_df['Y']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)

# Apply SMOTE for dealing with class imbalance
sm = SMOTE(random_state=42)
X_train_res, y_train_res = sm.fit_resample(X_train, y_train)

# Logistic Regression Classifier
classifier = LogisticRegression()

# Define hyperparameters for RandomizedSearchCV
param_grid = {
    'penalty': ['l1', 'l2'],
    'C': [0.001, 0.01, 0.1, 1, 10, 100]
}

# Perform RandomizedSearchCV
lr_random = RandomizedSearchCV(estimator=classifier, param_distributions=param_grid, scoring='f1', n_iter=100, cv=5, verbose=2, random_state=42, n_jobs=-1)
lr_random.fit(X_train_res, y_train_res)

# Get the best model from RandomizedSearchCV
best_lr_model = lr_random.best_estimator_

# Make predictions on the training and test sets
y_train_pred = best_lr_model.predict(X_train_res)
y_test_pred = best_lr_model.predict(X_test)

# Evaluate the model on both training and test sets
train_accuracy = accuracy_score(y_train_res, y_train_pred)
test_accuracy = accuracy_score(y_test, y_test_pred)

print("Training Accuracy:", train_accuracy)
print("Test Accuracy:", test_accuracy)

# Generate classification reports for both training and test sets
print("Training Classification Report:")
print(classification_report(y_train_res, y_train_pred))

print("Test Classification Report:")
print(classification_report(y_test, y_test_pred))
```

Ensemble:

```python
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import classification_report, accuracy_score
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import TomekLinks
import numpy as np

X = processed_df.drop('Y', axis=1)
y = processed_df['Y']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)

# Handle NaN values in y_test
nan_indices = np.isnan(y_test)
X_test = X_test[~nan_indices]
y_test = y_test[~nan_indices]

smote = SMOTE(k_neighbors=5, sampling_strategy='auto', random_state=42)
tomek_links = TomekLinks(sampling_strategy='majority')

X_train_res, y_train_res = smote.fit_resample(X_train, y_train)
X_train_res, y_train_res = tomek_links.fit_resample(X_train_res, y_train_res)

rf_classifier = RandomForestClassifier(
    n_estimators=1000,
    max_depth=100,
    min_samples_split=2,
    min_samples_leaf=4,
    random_state=42
)
xgb_classifier = XGBClassifier(
    n_estimators=394,
    colsample_bytree=0.7123738038749523,
    gamma=0.27134804157912423,
    subsample=0.9947547746402069,
    learning_rate=0.052277267492428794,
    max_depth=9,
    min_child_weight=1,
    random_state=42
)

ensemble_model = VotingClassifier(estimators=[
    ('random_forest', rf_classifier),
    ('xgboost', xgb_classifier)
], voting='soft')

ensemble_model.fit(X_train_res, y_train_res)

# Training result
y_pred_train = ensemble_model.predict(X_train_res)
print("Ensemble Model Training Performance:")
print(classification_report(y_train_res, y_pred_train))
print("Training Accuracy:", accuracy_score(y_train_res, y_pred_train))

# Test result
y_pred_test = ensemble_model.predict(X_test)
print("Ensemble Model Test Performance:")
print(classification_report(y_test, y_pred_test))
print("Test Accuracy:", accuracy_score(y_test, y_pred_test))
```

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid for grid search
param_grid = {
    'random_forest__n_estimators': [100, 500],
    'random_forest__max_depth': [50, 100],
    'xgboost__n_estimators': [200, 400],
    'xgboost__max_depth': [5, 10],
}

# Create the grid search object
grid_search = GridSearchCV(estimator=ensemble_model, param_grid=param_grid, scoring='accuracy')

# Perform grid search on the resampled training data
grid_search.fit(X_train_res, y_train_res)

# Get the best estimator and its associated parameters
best_estimator = grid_search.best_estimator_
best_params = grid_search.best_params_

# Print the best parameters
print("Best Parameters:", best_params)

# Training result with best estimator
y_pred_train = best_estimator.predict(X_train_res)
print("Ensemble Model Training Performance:")
print(classification_report(y_train_res, y_pred_train))
print("Training Accuracy:", accuracy_score(y_train_res, y_pred_train))

# Test result with best estimator
y_pred_test = best_estimator.predict(X_test)
print("Ensemble Model Test Performance:")
print(classification_report(y_test, y_pred_test))
print("Test Accuracy:", accuracy_score(y_test, y_pred_test))
```
